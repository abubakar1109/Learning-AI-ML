Visualization of Dataset

Histogram

```python
import matplotlib.pyplot as plt

data.hist(figsize=(10,10))
plt.show()
```

Density Plot

```python
data.plot(kind='density', subplots=True, sharex=False, layout=(3,3), figsize=(15,10))
```

Box and Whisker Plot

```python
data.plot(kind='box', subplots=True, sharex=False, label=(3,3), figsize=(15,10))
```

Correlation Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True)
```

Scatter Plot Matrix

```python
pd.plotting.scatter_matrix(data, figsize=(15,10))
plt.show()
```
